```python
import numpy as np
import math
from math import pi
import matplotlib.pyplot as plt, numpy as np, scipy.constants as ct
```

```python
freqs = np.geomspace(1e3, 1e9, 100, dtype=np.complex_)
omegas = 2 * math.pi * freqs
mu_0 = 4 * math.pi * 1e-7
sigma = 5e-3
epsilon = 8.85e-11
```

```python
print (2*math.pi*1e5* epsilon/sigma)
```

```python
k_vals = np.zeros(100, dtype = np.complex_)
for n in range(len(omegas)):
    int_term1 = complex(1, -(sigma/(omegas[n] * epsilon)))
    k_vals[n] = omegas[n] * ((mu_0 * epsilon * int_term1) ** (1/2))

k_pp = np.imag(k_vals)

pen_depths = np.zeros(100, dtype = np.complex_)
for n in range(len(k_vals)):
    pen_depths[n] = (math.log(0.01))/(2*k_pp[n])

```

```python
plt.figure(figsize = (6,4))
plt.plot(np.real(freqs), np.real(pen_depths))

plt.ylabel("Penetration Depth (m)")
plt.xlabel("Frequency (Hz)")
plt.xscale('log')
plt.title("Penetration Depth of Ground")
plt.show()
```
