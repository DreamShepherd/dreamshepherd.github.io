```python
phi, theta = np.linspace(0, 2*np.pi, 40), np.linspace(0, np.pi, 40)
arr_phi, arr_theta = np.meshgrid(phi, theta)
arr_r = np.abs((1 + np.cos(np.pi * np.cos(arr_theta) + np.pi))/2)
arr_x = arr_r * np.sin(arr_theta) * np.cos(arr_phi)
arr_y = arr_r * np.sin(arr_theta) * np.sin(arr_phi)
arr_z = arr_r * np.cos(arr_theta)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = '3d')
cmap = plt.get_cmap('jet')
plot = ax.plot_surface(arr_x, arr_y, arr_z, rstride = 1, cstride = 1, facecolors=cmap(arr_r), linewidth = 0, antialiased = False, alpha = 0.5)
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
ax.set_title("Radiation Pattern")
plt.show()
print (np.max(arr_r))
```
