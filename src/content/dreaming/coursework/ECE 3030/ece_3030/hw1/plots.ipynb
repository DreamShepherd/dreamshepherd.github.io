```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = x
v = y
w = z

ax.quiver(x, y, z, u, v, w, length=0.15)

plt.title("Graph 1")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")



plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.4, 0.4, 0.15),
                      np.arange(-0.4, 0.4, 0.15),
                      np.arange(-0.5, 0.5, 0.2))

r = (x ** 2 + y ** 2 + z ** 2) ** (3/2)
u = x / r
v = y / r
w = z / r

ax.quiver(x, y, z, u, v, w, length=0.005)

plt.title("Graph 2")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")



plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-3,3,15),np.linspace(-3,3,15))

u = x/(np.sqrt(x**2 + y**2) ** (3))
v = y/(np.sqrt(x**2 + y**2) ** 3)

plt.quiver(x,y,u,v)
plt.title("Graph 3")
plt.xlabel("$x$")
plt.ylabel("$y$")


plt.show()
```
